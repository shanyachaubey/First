```python
#Insertion sort

def insertion_sort(list):
    ordered_list = []
    for i in range(0, len(list)):
        ordered_list.append(list[i])
        j = len(ordered_list)-1
        while j>0 and ordered_list[j] < ordered_list[j-1]:
            ordered_list[j], ordered_list[j-1] = ordered_list[j-1], ordered_list[j]
            j -= 1
    return ordered_list



list = [1,2,3,4,34,21,0]
print(insertion_sort(list))
```

```python
def selection_sort(list):
    ol = []
    for i in range(0, len(list)):
        m = find_min(list, i ,len(list))
        list[m], list[i] = list[i], list[m]
        ol.append(list[i])
    return ol



print(selection_sort([9,2,1,54,0, 0, 0]))
```

```python
def bubble_sort3(list):
 #’’’This bubble_sort does the sorting in-place.
 #It takes a list and swap biger elements to the right until no change needed.’’’

    print("Bubble sort for", list, "starts.")
    n = len(list)
    while True:
        swap = False
        for i in range(n - 1):
            if i == n - i:
                break
            if list[i] > list[i+1]:
                list[i], list[i+1] = list[i+1], list[i]
                swap = True
            print("partially ordered list as:", list)
            if swap == False:
                break
        return list
```
